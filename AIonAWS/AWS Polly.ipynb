```python
import boto3
po = boto3.client('polly')
res = po.synthesize_speech(Text="Hello guys, I love you all", OutputFormat='mp3', VoiceId = 'Joanna')
```

```python
#res['AudioStream']
file = open('myaudio.mp3', 'wb')
```

```python
file.write(res['AudioStream'].read())
file.close()
```
